```
import pendulum
```

```
df = pd.read_excel(DATA_DIR + 'covid19_detect.xlsx')
df
```

```
column = ['日期', '檢驗人數', '確診人數', '居檢送驗', '居檢送驗累計', '武肺通報', '武肺通報累計', '擴大監測','擴大監測累計']
df = df[column]
```

```
df.fillna(value={col: 0 for col in df.columns[1:]}, inplace=True)
for col in df.columns[1:]:
    df[col] = df[col].astype(int)
```

```
print(
    '送驗:'+ str(df['檢驗人數'].sum()),
    '確診:'+ str(df['確診人數'].sum()),   
)
```

```
yesterday_dt = pendulum.datetime(2020, 6, 12).format("YYYY-MM-DD")
df[df['日期'] == yesterday_dt].loc[:, ['日期', '檢驗人數', '確診人數']]\
    .rename({'檢驗人數': "送驗", '確診人數': '確診'}, axis=1)
```

```
data = (
    df.rename({ 
        '擴大監測': '擴大監測送驗',
        '居檢送驗': '居家檢驗送驗',
        '武肺通報': '法定傳染病通報',
    }, axis=1) 
    .melt('日期', ['擴大監測送驗', '居家檢驗送驗', '法定傳染病通報'], 
          var_name='通報來源', value_name="通報數")
    .fillna(0)
)
```

```
import altair as alt 
_ = alt.renderers.set_embed_options(actions=False)

chart_title = "COVID-19（武漢肺炎）-通報來源"
alt.Chart(data, title=chart_title).mark_bar().encode(
    x=alt.X('日期:T', title="通報日", axis=alt.Axis(
        grid=False,
        format="%m/%d",
    )),y="通報數:Q",
    color='通報來源:N',
    tooltip=['日期:T', '通報來源', '通報數']
).properties(
    width=1200
)
```

```
df2 = pd.read_excel(DATA_DIR + 'covid19_test.xlsx')
df2
```

```
df2 = df2[:443]
df2 = df2[['新聞稿發布日期', '案例', '來源', '性別', '年齡']]
```

```
data = (
    df2
    .groupby(['新聞稿發布日期', '來源'])['案例']
    .count()
    .reset_index()
)
```

```
title = "確診個案趨勢圖 - 依記者會日期"
alt.Chart(data, title=title).mark_bar().encode(
    x=alt.X('新聞稿發布日期:T', title="發病日", axis=alt.Axis(
        grid=False,
        format="%m/%d")),
    y='案例:Q',
    color='來源:N',
    tooltip=['新聞稿發布日期:T', '來源', '案例']
).properties(
    width=1200
)
```
